You can customize the default world map by setting its center and initial zoom level.

Maps are defined by Latitude and Longitude values. For example, a center at [0, 0] uses those coordinates.

You can also set the initial zoom level for a defined center. A higher zoom level means a closer view of the center.

Let's create a map centered around Canada and adjust the zoom level to see the changes.

```python
import folium

# Define a function to create and display a map centered around Canada
def display_map(center, zoom_level):
    # Create the map
    world_map = folium.Map(location=center, zoom_start=zoom_level)
    # Display the map
    return world_map

# Coordinates for Canada
canada_center = [56.130, -106.35]

# Display the map with a higher zoom level
high_zoom_map = display_map(canada_center, 8)

# Display the map with a lower zoom level
low_zoom_map = display_map(canada_center, 4)


```

**Maps with Markers**

Download the dataset and load it into a pandas DataFrame:

```python
df_incidents = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Police_Department_Incidents_-_Previous_Year__2016_.csv')

print('Dataset downloaded and read into a pandas dataframe!')
```

```python
df_incidents.head()
```

Each row in the dataset contains 13 features:

- **IncidntNum**: Incident Number
- **Category**: Category of crime or incident
- **Descript**: Description of the crime or incident
- **DayOfWeek**: Day of the week when the incident occurred
- **Date**: Date when the incident occurred
- **Time**: Time of day when the incident occurred
- **PdDistrict**: Police department district
- **Resolution**: Resolution of the crime (e.g., whether the perpetrator was arrested)
- **Address**: Closest address to the incident location
- **X**: Longitude of the crime location
- **Y**: Latitude of the crime location
- **Location**: Tuple of latitude and longitude values
- **PdId**: Police department ID

Let's determine how many entries are in our dataset.

```python
df_incidents.shape
```

The DataFrame contains 150,500 incidents from the year 2016. To reduce computational cost, let's work with only the first 100 incidents from this dataset.

```python
# Get the first 100 incidents from the df_incidents DataFrame

df_incidents = df_incidents.head(100)
```

```python
df_incidents.shape
```

Now that we've reduced the data, let's visualize the locations of these crimes in San Francisco. We will use the default map style and set the zoom level to 12.

```python
# San Francisco latitude and longitude values
latitude = 37.77
longitude = -122.42
```

```python
# Create the map and set the initial zoom level to 12
sanfran_map = folium.Map(location=[latitude, longitude], zoom_start=12)

# Display the map
sanfran_map
```

Next, we'll overlay the crime locations on the map. In Folium, you do this by creating a feature group with its own features and style, and then adding it to `sanfran_map`.

```python
incidents = folium.FeatureGroup()

# Loop through the 100 crimes and add each to the incidents feature group
for lat, lng in zip(df_incidents.Y, df_incidents.X):
    incidents.add_child(
        folium.CircleMarker(
            location=[lat, lng],
            radius=5,  # Define the size of the circle markers
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        )
    )

# Add the incidents feature group to the map
sanfran_map.add_child(incidents)
```

You can also add pop-up text that appears when you hover over a marker. Let’s configure each marker to display the crime category when hovered over.

```python
# Instantiate a feature group for the incidents
incidents = folium.FeatureGroup()

# Loop through the 100 crimes and add each as a CircleMarker to the incidents feature group
for lat, lng, label in zip(df_incidents.Y, df_incidents.X, df_incidents.Category):
    incidents.add_child(
        folium.CircleMarker(
            location=[lat, lng],
            radius=5,  # Define the size of the circle markers
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6,
            popup=folium.Popup(label, parse_html=True)  # Add pop-up text to each marker
        )
    )

# Add the incidents feature group to the map
sanfran_map.add_child(incidents)
```

Isn’t this cool? Now you can see the crime category for each marker.

If the map feels congested with all these markers, you have two options. The simpler solution is to remove the separate location markers and add the text directly to the circle markers instead.

```python
# Loop through the 100 crimes and add each as a CircleMarker to the map
for lat, lng, label in zip(df_incidents.Y, df_incidents.X, df_incidents.Category):
    folium.CircleMarker(
        location=[lat, lng],
        radius=5,  # Define the size of the circle markers
        color='yellow',
        fill=True,
        fill_color='blue',
        fill_opacity=0.6,
        popup=folium.Popup(label, parse_html=True)  # Add text directly to the marker
    ).add_to(sanfran_map)

# Display the map
sanfran_map
```

Another effective approach is to group the markers into clusters. Each cluster represents the number of crimes in a specific neighborhood. These clusters can be viewed as distinct areas of San Francisco for more detailed analysis.

To implement this, start by creating a `MarkerCluster` object and adding all the data points from the DataFrame to it.

```python
from folium import plugins
# Instantiate a MarkerCluster object
marker_cluster = plugins.MarkerCluster().add_to(sanfran_map)

# Loop through the DataFrame and add each data point to the MarkerCluster
for lat, lng, label in zip(df_incidents.Y, df_incidents.X, df_incidents.Category):
    folium.Marker(
        location=[lat, lng],
        popup=label
    ).add_to(marker_cluster)

# Display the map
sanfran_map
```
